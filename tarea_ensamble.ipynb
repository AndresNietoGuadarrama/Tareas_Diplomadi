# Ejemplo de uso de baggings y ensamble

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
import sklearn.compose
from sklearn.datasets import make_moons
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import f1_score
from numpy.random import seed, randn
from sklearn.preprocessing import PolynomialFeatures
```

### Datos originales

```python
noise = 0.2
xtrain, ytrain = make_moons(200, noise=noise, random_state=2718)
xtest, ytest = make_moons(40, noise=noise,random_state=31415)
plt.subplot(2,1,1)
plt.scatter(*xtrain.T, c=ytrain, cmap="Set2",label="train")
plt.legend()
plt.subplot(2,1,2)
plt.scatter(*xtest.T, c=ytest, cmap="Set2",label="test")
plt.legend()
plt.suptitle("Datos originales")
plt.show()
```

### 1. Creación del pipeline con un modelo polinomial y de regresión logíca

```python
deg=30
model=Pipeline([
        ("Polytransform",PolynomialFeatures(degree=deg,include_bias=False)),
        ("log_reg",LogisticRegression(penalty="none",solver="newton-cg"))
])
```

### 2. Bagging

```python
from sklearn.ensemble import BaggingClassifier
n_est,random_st=100,314
bagging_logreg = BaggingClassifier(model, n_estimators=n_est, random_state=random_st)
```

### 3. Entrenamiento de los dos modelos

```python
model.fit(xtrain,ytrain)
bagging_logreg.fit(xtrain,ytrain)
```

### 4. Graficación de función de decisión

```python
import plot_classf_model
from plot_classf_model import plot_classf_model
fig, ax = plt.subplots(1, 2, figsize=(12, 4))
plot_classf_model(xtrain,ytrain,model, ax=ax[0], contour=True,title="Función de decisión de modelo sin bagging")
plot_classf_model(xtrain,ytrain,bagging_logreg, ax=ax[1], contour=True,title="Función de decisión de modelo con bagging")
```

Observar que el modelo con bagging al parecer clasifica mejor los datos. Esto debería reflejarse en un score $F_1$ más alto para el modelo con bagging que para el modelo por separado.

# 5. $F_{1}$ score

```python
tipos=['macro','micro','weighted']
ypred_model=model.predict(xtest)
ypred_bagging=bagging_logreg.predict(xtest)
for t in tipos:
    print(f"Score F1 del modelo sin bagging y con {t} avarage es: {round(f1_score(ytest,ypred_model, average=t),3)}")
    print(f"Score F1 del modelo con bagging y con {t} avarage es: {round(f1_score(ytest,ypred_bagging, average=t),3)}")
    if t != 'weighted':
        print("-----------------------------------------------------------------")
```
